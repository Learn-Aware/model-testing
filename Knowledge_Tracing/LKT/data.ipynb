```python
import pandas as pd

data = pd.read_csv('data/assistments_2009_2010/assistments_2009_2010.csv')

data.head()
```

```python
# heatmap showing NaN values
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(data.isnull(), cbar=False)
```

```python
# drop rows with NaN values
data = data.dropna()

data
```

```python
print(data['answer_type'].unique().tolist())
print(data['tutor_mode'].unique().tolist())
print(data['list_skills'].unique().tolist())
print(data['correct'].unique().tolist())

```

```python
## Template dataset creation
columns = ['user_id', 'template']

'''
Template:
Here is a student’s problem-solving history:
((list_skills, answer_type,) correct)

'''
```

```python
df = data.copy()
# Function to create the template string
def create_template(row):
    return f"(({row['list_skills']}, {row['answer_type']}), {float(row['correct'])})"

# Create a new dataframe with 'user_id' and 'template' columns
df_template = df[['user_id']].copy()
df_template['template'] = df.apply(create_template, axis=1)

df_template
```

```python
# get users with most templates
df_template['user_id'].value_counts().head()
```

```python
df_student = df_template.groupby('user_id')['template'].apply(lambda x: ' -> '.join(x)).reset_index()

df_student
```
